```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

# Carrega o dataset CIFAR-10
cifar10 = fetch_openml('cifar_10', version=1, cache=True)

# Separa os dados e os rótulos
X = cifar10.data
y = cifar10.target

# Converte os rótulos para inteiros
y = y.astype(np.uint8)

```

```python
# Divide os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train, y_train = X_train[:24000], y_train[:24000]

# Normaliza os dados
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
from sklearn import ensemble
```

```python
randomF = ensemble.RandomForestClassifier(random_state=1)
```

```python
randomF.fit(X_train, y_train)
```

```python
from sklearn import linear_model
```

```python
passiveagC = linear_model.PassiveAggressiveClassifier(random_state=1, max_iter=100)
```

```python
passiveagC.fit(X_train, y_train)
```

```python
sgdC = linear_model.SGDClassifier(random_state=1, max_iter=100)
```

```python
sgdC.fit(X_train, y_train)
```

```python
from sklearn import metrics
```

```python
esperado1 = y_test
predito1 = randomF.predict(X_test)
print('Relatório de Classificação: %s \n %s \n' % (randomF, metrics.classification_report(esperado1, predito1)))
```

```python
esperado2 = y_test
predito2 = passiveagC.predict(X_test)
print('Relatório de Classificação: %s \n %s \n' % (passiveagC, metrics.classification_report(esperado2, predito2)))
```

```python
esperado3 = y_test
predito3 = sgdC.predict(X_test)
print('Relatório de Classificação: %s \n %s \n' % (sgdC, metrics.classification_report(esperado3, predito3)))
```
